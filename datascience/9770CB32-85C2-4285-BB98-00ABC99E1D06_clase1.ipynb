## Medida de tendencia central Media

$$
\Large{
\bar{x} = \frac{1}{n}\sum_{i=1}^{n}x_i
}
$$

```python
# python nativo
prom1 = sum(df['altura']) / len(df['altura'])

print(f'Promedio altura: {prom1}')
```

```python
# pandas
prom2 = df['altura'].mean()

print(f'Promedio altura: {prom2}')
```

```python
# numpy
prom3 = np.mean(df['altura'])

print(f'Promedio altura: {prom3}')
```

## Varianza

- Varianza poblacional
$$
\Large{
\sigma^2 = \frac{\sum_{i=1}^{n}(x_i-\mu)^2}{N}
}
$$


- Varianza muestral
$$
\Large{
S^2 = \frac{\sum_{i=1}^{n}(x_i-\bar{x})^2}{N-1}
}
$$



```python
# varianza poblacional
mu = df['altura'].mean()

var_pob = sum((df['altura']-mu)**2)/len(df['altura'])
var_pob
```

```python
np.var(df['altura'])
```

```python
# este calculo es de la varianza muestral
print(df['altura'].var())


# varianza poblacion usando pandas
print(df['altura'].var(ddof=0))
```

```python
# varianza muestral con numpy
np.var(df['altura'], ddof=1)
```

```python
df['altura'].mean()
```

### Desviacin estndar

- Std poblacional
$$
\Large{
\sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i-\mu)^2}{N}}
}
$$


- Std muestral
$$
\Large{
S = \sqrt{\frac{\sum_{i=1}^{n}(x_i-\bar{x})^2}{N-1}}
}
$$


```python
# desviacin estndar poblacional en numpy
np.std(df['altura'])
```

```python
# desviacin estndar muestral en numpy
np.std(df['altura'], ddof=1)
```

```python
# desviacin estndar muestral
print(df['altura'].std())

# desviacin estndar poblacional
print(df['altura'].std(ddof=0))
```
