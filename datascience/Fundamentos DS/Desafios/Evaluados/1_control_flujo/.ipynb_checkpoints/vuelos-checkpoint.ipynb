```python
import pandas as pd
import numpy as np
```

#### 1. Generar arrays ficticios

```python
array1 = np.linspace(1,50,50)
array2 = np.linspace(50,150,101)
print(array1)
print(array2)
```

#### 2. Loop que devuelve si un número es par o impar

```python
for n in array1:
    if n%2 == 0:
        print(f'El número {n} es par')
    else:
        print(f'El número {n} es impar')
```

#### 3. Loop que devuelve si un número es:
- Divisible por 2 o 3
- Divisible por 2 y 3
- Divisible por 3 y no por 2
- No es divisible por 2 ni por 3

```python
for n in array2:
    if n%2 == 0 and n%3 == 0:
        print(f'{n} es divisible por 2 y 3')
    elif n%2 != 0 and n%3 == 0:
        print(f'{n} es divisible por 3 pero no por 2')
    elif n%2 == 0 or n%3 == 0:
        print(f'{n} es divisible por 2 o 3')
    else:
        print(f'{n} no es divisible por 2 ni 3')
```

#### 5. Utilizando la misma base flights.csv debes hacer un loop y clasificar los meses con una cantidad de pasajeros menor a la media. 

```python
#Se carga el archivo flights.csv
df = pd.read_csv('flights.csv')

#Se obtiene la media de la cantidad de pasajeros
media_pax = np.mean(df['passengers'])

#Se crea nueva columna underperforming
df['underperforming'] = 0

#Recorrer Passengers y asignar un 1 a la columna underperforming cuando corresponde
for i, serie in df.iterrows():
    if serie['passengers'] < media_pax:
        df.at[i,'underperforming'] = 1

df
```

#### 6. Loop que clasifique los meses donde la cantidad de pasajeros se escapa de la tendencia. 

```python
#Se obtiene media general de los pasajeros
media_pax = np.mean(df['passengers'])

#Se obtiene la desviacion estandar de los pasajeros
std_pax = np.std(df['passengers'])

#Nueva columna outlier con 0 por defecto
df['outlier'] = 0

#Loop que recorra cada observación de passengers, donde si la
#observación se escapa de la tendencia sea 1, de lo contrario 0.
casos_extremos = 0
for i, serie in df.iterrows():
    if serie['passengers'] < media_pax-std_pax or serie['passengers'] > media_pax+std_pax:
        df.at[i,'outlier'] = 1
        casos_extremos += 1
        
print(f'Casos extremos: {casos_extremos}')
df[df['outlier'] == 1]
```
