#### 2. Importar la base de datos utilizada la sesion presencial anterior

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import warnings

warnings.filterwarnings(action='ignore')
plt.style.use('seaborn')

df = pd.read_csv('nations.csv', encoding='iso-8859-1')
df
```

```python
#DataFrame con los datos a considerar
subset = df[['region','femlab','literacy','school']].dropna()
subset
```

```python
for region in subset['region'].unique():
    subset[f'is_{region}'] = np.where(subset['region'] == region, 1, 0) 
subset
```

#### 3. Implementar una función de prueba de hipótesis a mano:

```python
#stats.ttest_ind()
#Se entregan 2 muestras, correspondientes a las muestras de las dos poblaciones a comparar
#Se debe separar el df en dos grupos segun la variable que se indique dentro del parametro binarize

def hipotesis(df,variable,binarize):
    var1 = df[variable][df[binarize] == 1].dropna()
    var0 = df[variable][df[binarize] == 0].dropna()
    
    t,pval = stats.ttest_ind(var1,var0)
    
    print(f'Media de {variable} para {binarize} = 1: {var1.mean()}')
    print(f'Media de {variable} para {binarize} = 0: {var0.mean()}')
    print(f'La diferencia entre las medias es de: {var1.mean() - var0.mean()}')
    print(f'Valor t = {t}')
    print(f'Valor p = {pval}')
    
    return t,pval
    
hipotesis(subset,'school','is_Americas')
```

#### 4. Implementar una función que grafique los histogramas para ambas muestras

```python
def graph_hist(df,variable,binarize):
    var1 = df[variable][df[binarize] == 1].dropna()
    var0 = df[variable][df[binarize] == 0].dropna()
    
    sns.histplot(data=df, x=var0, color='skyblue', label=f'{variable} para {binarize} = 0', kde=True)
    sns.histplot(data=df, x=var1, color='red', label=f'{variable} para {binarize} = 1', kde=True)
    plt.legend() 
```

##### Continente Europeo

```python
#Promedio años de escolaridad
hipotesis(subset,'school','is_Europe')
graph_hist(subset,'school','is_Europe')
```

- La media en el promedio de años de escolaridar en Europa es superior versus al resto del mundo llegando a 9.98 años. Mientras que, en el resto del mundo se observa una media de 6 años.

```python
#Tasa de alfabetismo
hipotesis(subset,'literacy','is_Europe')
graph_hist(subset,'literacy','is_Europe')
```

- La media de la tasa de alfabetismo en el continente Europeo es claramente superior al resto del mundo llegando a un 98.2%. Mientras que, la tasa de alfabetismo del resto del mundo tiene una media de 77.7%

```python
#Tasa entre hombres y mujeres en el mercado laboral
hipotesis(subset,'femlab','is_Europe')
graph_hist(subset,'femlab','is_Europe')
```

- La media en la tasa entre hombres y mujeres en el mercado laboral en Europa es similar a la media en el resto del mundo, obteniendo una diferencia de 0.08% en las medias
