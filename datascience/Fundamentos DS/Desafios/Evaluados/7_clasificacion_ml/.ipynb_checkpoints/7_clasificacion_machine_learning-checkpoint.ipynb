### Desafio 2: Análisis exploratorio
Explore las principales relaciones entre las variables. Utilizando los métodos boxplot y lmplot de seaborn implemente los siguientes gráficos:
- Un boxplot del balance total de las cuentas, segmentado por el status del estudiante.
- Un boxplot del ingreso total de cada cliente, segmentado por el status de estudiante.
- Un boxplot del balance total de las cuentas, segmentado por el status de default.
- Un boxplot del ingreso total de cada cliente, segmentado por el status de default.
- Un diagrama de dispersión entre balance e income, segmentado por el status de estudiante.
- Un diagrama de dispersión entre balance e income, segmentado por el status de default.

```python
#Funcion para graficar variables del DF con boxplot
def graficar_boxplot(df):
    plt.rcParams['figure.figsize'] = (10,10)
    cat = []
    num = []
    for n,i in enumerate(df):
        if (df[i].dtypes == 'object'):
            cat.append(i)
        else:
            num.append(i)
    print(f'Variables numéricas: {num}')
    print(f'Variables categóricas: {cat}')
    n=1
    for i in cat:
        for j in num:
            plt.subplot(int(len(list(df.columns))/3)+2,3,n)
            ax = sns.boxplot(x=i,y=j,data=df)
            n+=1
    plt.tight_layout()
```

```python
#Se usa la función df_boxplot() para graficar los boxplot
graficar_boxplot(df)
```

- En cuanto a la morosidad, podemos observar que el balance de la cuenta de crédito es superior para los clientes morosos. Mientras que, en relación a los ingresos no se observa una diferencia significativa entre morosos y no morosos.
- En cuanto a los estudiantes, podemos observar en el balance un pequeña diferencia entre morosos y no morosos. Mientras que, en cuanto al ingreso observamos que en los estudiantes es significantemente menor que los no estudiantes.

```python
#Se genera diagrama de dispersión entre balance e income segmentado por el status de estudiante.
sns.lmplot(x="balance",y="income",hue="student",data=df)
plt.tight_layout()
```

- Esto nos vuelve a demostrar que los estudiantes tienen un ingreso inferior vs los no estudiantes.

```python
#Se genera diagrama de dispersión entre balance e income segmentado por el status de default.
sns.lmplot(x="balance",y="income",hue="default",data=df)
plt.tight_layout()
```

- Se observa que en la medida que aumenta el balance es más probable caer en default, independientemente del income.
- Tambien observamos que la relación entre balance e income es inversamente proporcional para cuando cae y no cae en default, siendo esta última en menor proporción.

### Desafío 3: Modelación econométrica

- Recodifique las variables student y default para que adquieran atributos numéricos

```python
#Se hace una copia del df original
df_recod = df.copy()
#Se refactoriza student
df_recod['student'] = np.where(df_recod['student'] == 'Yes',1,0)
#Se refactoriza default
df_recod['default'] = np.where(df_recod['default'] == 'Yes',1,0)
#Se revisa df_recod
df_recod.head()
```

```python
df_recod.info()
```

- Implemente el siguiente modelo econométrico:
$$\log \left(\frac{Pr(default=1)}{1-Pr(default=1)}\right)=\beta_{0} + \beta_{1} \times balance + \beta_{2} \times income + \beta_{3} \times student$$
- A partir de este modelo, depúrelo y mantenga las principales características. Ejecute de nuevo el modelo y comente brevemente sobre el comportamiento de los atributos.

```python
#Se implementa el modelo saturado
m_logit = smf.logit('default ~ balance + income + student', df_recod).fit()
```

```python
#Se revisa el modelo saturado
m_logit.summary2()
```

- Podemos observar que la columna tiene un p_value de 0.7115, como es muy superior a 0.005 no se considera en el modelo depurado que se implementa a continuación:

```python
#Se implementa el modelo depurado
m_logit_dep = smf.logit('default ~ balance + student', df_recod).fit()
```

```python
#Se revisa el modelo depurado
m_logit_dep.summary2()
```

- Se puede ver que el valor del Pseudo R cuadrado se mantiene.
- Tambien se observa que la variable balance afecta positivamente con el aumento en la chance de que la persona entre en morosidad. Mientras que, la variable student afecta negativamente en la chance de que la persona entre en morosidad

### Desafío 4
A partir del modelo depurado, genere un modelo predictivo con las siguientes características:
- Genera subconjuntos de entrenamiento y pruebas, separando el 33% de la muestra en el conjunto de pruebas.
- Estandarice las matrices de atributos de entrenamiento y pruebas usando StandardScaler() ajustado solo sobre el conjunto de entrenamiento.
- Implemente el modelo sin alterar los hiper parámetros.
- Genere la predicción de las clases y de la probabilidad.
- Reporte las métricas de precisión, recall y f1 para ambas clases.
- Grafique la curva Receiver Operator Characteristic
- En base a los pasos, comente sobre el desempeño predictivo del modelo haciendo mención a todas las métricas calculadas y a la curva ROC.

```python
#Se generan lo subsets para la variable dependiente y la variable independiente
df_y = df_recod['default']
df_x = df_recod.drop(columns=['default','income'])
```

```python
#Se generan los subconjuntos de entrenamiento y validación separando el df según el porcentaje entregado
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y, test_size=0.33, random_state = 13579)
```

```python
#Se estandarizan las matrcies de atributos de entrenamiento y pruebas con StandardScaler() ajustando
#solo el conjunto de entrenamiento
x_train_std_media = StandardScaler().fit(x_train)
x_train_std = x_train_std_media.transform(x_train)
```

```python
x_train.head()
```

```python
#Se implementa modelo sin alterar los hiperparámetros
default_machine_model = LogisticRegression(random_state=18576).fit(x_train_std,y_train)
```

```python
#Se revisan los coeficientes
default_machine_model.coef_
#array([[-0.31713005,  2.64794615]])
```

- Se ve que los coeficientes entregados por el modelo default_machine_model son diferentes a los entregados por el modelo econométrico

```python
x_test_std_media = StandardScaler().fit(x_test)
x_test_std = x_test_std_media.transform(x_test)
```

```python
#Se generan predicciones en base a nuestra matríz de prueba
yhat = default_machine_model.predict(x_test_std)
yhat[:10]
```

```python
#Se generan las predicciones como probabilidad
yhat_prob = default_machine_model.predict_proba(x_test_std)
yhat_prob[:10].round(2)
```

```python
m_accuracy = accuracy_score(y_test,yhat)
m_accuracy.round(4)
```

- Se observa que este modelo tiene una exactitud del 97.79%

```python
#Se obtienen las métricas de precisión, recall y F1 para ambas clases
print(classification_report(y_test,yhat))
```

- Se puede observar que hay una presición de:
    - 98% para la clase 0
    - 77% para la clase 1
- Se puede observar que hay una sensibilidad (recall) de:
    - 100% para la clase 0
    - 34% para la clase 1
- Se puede observar que hay una media armónica entre presición y sensibilidad (F1-Score) de:
    - 99% para la clase 0
    - 47% para la clase 1

```python
#Se grafica la curva ROC
plt.rcParams['figure.figsize'] = (12,10)
#Se estima nuevamente los valores para obtener la probabilidad entre 0 y 1
yhat_roc = default_machine_model.predict_proba(x_test_std)[:,1]
#Se generan los objetos de roc_curve
false_positive,true_positive,threshold = roc_curve(y_test,yhat_roc)
```

```python
#Se crea el gráfico
plt.title('Curva ROC')
plt.plot(false_positive, true_positive, lw=1)
plt.plot([0,1], ls='--', lw=1)
plt.plot([0,0], [1,0], c='limegreen', lw=3)
plt.plot([0,1], [1,1], c='limegreen', lw=3)
```

```python
#Se obtiene el área bajo la curva
roc_auc_score(y_test,yhat_roc)
```

- Se puede observar que hay un desempeño predictivo de un 45.15% superior a una clasificación aleatoria.
